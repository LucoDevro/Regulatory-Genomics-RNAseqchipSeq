The list of significantly differently expressed genes (`genes.diff.txt`) was supplied to gProfiler to query the Drosophila melanogaster GO database. All ambigious GO terms were assigned to the ones with the most hits. The figure and the table data were exported.

```bash
cut -d ',' -f 1,2,4 gProfiler.csv | tr ',' '\t' | tr -d '"' > enriched.func.txt
```

```bash
cat enriched.func.txt
```

Knowing that the Ir75 genes encode chemosensory receptors, it is not surprising to find GO terms both for neuronal functions ((cat)ion transporters and activity, cell communication...) and for sensory functions (transducer activity, receptor activity, signalling, stimulus response...).

Let's now differentiate between up- and downregulated genes (`down.genes.text` & `up.genes.txt`)

```bash
cut -d ',' -f 1,2,4 gProfiler.up.csv | tr ',' '\t' | tr -d '"' > enriched.func.up.txt
```

```bash
cat enriched.func.up.txt
```

Next to the same neuronal and sensory functional annotations, folding chaperones and unfolded and misfolded binding activities are prominent at the top of the list. Perhaps, the run-through transcripts are being translated to some extent to faulty fusion proteins, whom the chaperones have difficulties to fold, resulting in mis- or unfolded proteins.

```bash
cut -d ',' -f 1,2,4 gProfiler.down.csv | tr ',' '\t' | tr -d '"' > enriched.func.down.txt
```

```bash
cat enriched.func.down.txt
```

Downregulated genes, again, are responsible for sensory and neuronal functions, reflecting the loss of Ir75a and Ir75b phenotypes.

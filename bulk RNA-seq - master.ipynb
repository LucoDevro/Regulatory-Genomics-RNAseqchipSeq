The olfactory receptors are fundamental to the perception of odours in the brain. In many phyla, each neuron of the olfactory system expresses exactly one receptor gene of a large gene pool contained within the genome. The process regulating this precise expression in both vertebrates and invertebrates is thought to rely on some transcriptional activators of the target locus, and dito repressors and/or inaccessible chromatin structure of other loci.

A non-negligible part of this gene pool, however, consists of tandem gene arrays, which in case of animals and many other vertebrates spans almost the entire olfactory gene pool. The advantage and the precise mechanism of this tandem organisation is still unclear. The model system at hand is the Ir75 olfactory gene array (from 5' end to 3' end: Ir75c-Ir75b-Ir75a) in antennae of *Drosophila melanogaster*. Due to an unknown regulatory network, the Ir75 genes encode different receptor proteins for spatially distinct groups of sensory neurons.

In this study, the expression of the gene array is analysed using RNAi of the E93 protein. It has already been shown earlier that disabling E93 results in a drastic drop of Ir75a and Ir75b protein levels, while Ir75c is present in many additional neuron groups instead of only one. The result is that all Ir75 neurons express the C-type receptor. This could indicate E93 represses Ir75c and promotes expression of Ir75a and Ir75b, although it is in both cases not known whether that would happen via a direct or an indirect mechanism. The dataset at hand (GEO accession GSE150296) comprises the bulk RNA sequencing of the antennae of three E93RNAi animals and three control animals.

*Reference:* Mika K, Cruchet S, Chai PC, et al. Olfactory receptor-dependent receptor repression in Drosophila. Sci Adv. 2021;7(32):eabe3745. Published 2021 Aug 6. doi:10.1126/sciadv.abe3745

**QC & FastQC**

FastQC was performed after downloading all six replicates using `fastq-dump`. The prior and posterior inspections did not reveal any serious problems. For more detailed information, see the 'mapping & counting' notebook.

**Mapping**

After creating a genome directory for the STAR algorithm using the *Drosophila* reference genome FASTA file and annotations GTF file, all six samples were mapped to the reference genome using STAR. The mapping was successful. For the mapping statistics, see the 'mapping & counting' notebook. 

Visualising the mapped reads in IGV yields some remarkable observations. It seems transcription levels of Ir75a and Ir75b have not diminished at all, despite the absence of their respective receptor protein phenotypes. Moreover, quite some reads contain pieces of exons of two neighbouring genes, suggesting an intertwining of the genes of this array, i.e., transcription does not terminate at the stop codons between genes.

**Feature counting and differential expression**

All three Ir75 genes surprisingly appear significantly upregulated, despite the absence of their phenotypes. One possible explanation is that there are fused transcripts comprising exons of multiple genes. These run-through transcripts could inhibit transcription of the downstream genes Ir75a and Ir75b by transcriptional interference. Each time a read is counted for being linked to one of the separate genes, there is a fair chance it actually is a read of the interfering run-through transcript. Feature counter algorithms cannot distinguish between reads from run-through transcripts and from normal ones, as the reads are too short. Next-generation sequencing could elucidate this because of its longer read lengths.

Looking back at the IGV visualisations, one can notice that the levels of Ir75a and Ir75b reads are higher than the Ir75c reads level in the control samples, while in the RNAi samples, they all are at a comparable level. This suggests that transcription does not initiate at Ir75a or Ir75b in the RNAi case due to interference and that many reads could be part of a run-through transcript initiating at Ir75c.

**Functional analysis**

A functional analysis of the significantly different expressed genes using gProfiler revealed functional enrichment in sensory and neuronal activities, which is not surprising for olfactory sensory neurons. Remarkably, the most prominently upregulated genes are associated with folding chaperones and un- or misfolded proteins. Are the run-through transcripts being translated into faulty proteins?

### Genome mapping with STAR

```bash
mkdir -p mapping
cd mapping
```

Let's first generate the genome directory.

- `Genome.fna` is the RefSeq Genome of *Drosophila melanogaster*: assembly Release 6 plus ISO1 MT
- `Ann.gtf` is the annotated RefSeq Genome

Both were retrieved from the NCBI Genome database from this link at 24th Oct:
https://ftp.ncbi.nlm.nih.gov/genomes/all/GCF/000/001/215/GCF_000001215.4_Release_6_plus_ISO1_MT/

```bash
mkdir -p genome_db
STAR --runMode genomeGenerate \
     --runThreadN 8 \
     --genomeDir ./genome_db \
     --genomeFastaFiles ../Genome.fna \
     --sjdbGTFfile ../Ann.gtf
```

**Counting**

```bash
cd ..
ls -l *.gtf
```

Sequencing run-through transcripts can yield reads that overlap multiple genes. Therefore, it can be useful to count multi-overlapping reads using the `-O` flag. To avoid counting too much ambiguous genes because of this, I'll exploit the paired end information (flagged through `-p`) and make a read required to be mapped at both ends using the `-B` flag. As such, only reads that are more credible to overlap genes, are counted.

A quality score of 10 (`-Q 10`) can assist to filter out artefacts, although its impact may be very limited given that the FastQC reported very good overall read quality scores.

```bash
featureCounts \
    -p -O -B \
    -Q 10 \
    -g gene \
    -a Ann.gtf \
    -o counts \
    ./mapping/E1.bam ./mapping/E2.bam ./mapping/E3.bam ./mapping/C1.bam ./mapping/C2.bam ./mapping/C3.bam
```

About 91 % of all reads were successfully assigned in all samples. This looks OK to me.
